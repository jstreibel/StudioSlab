```python
from sympy import *
init_printing()
```

```python
r, alpha, eps, sigma = symbols("r alpha epsilon sigma")
```

```python
U = lambda r: eps/alpha * (1-r/sigma)**alpha * Heaviside(1-r/sigma)
U(r)
```

```python
V = U(r).subs(((sigma, 1), (eps, 1), (alpha, 0.001)))
dV = -V.diff(r)
plot(V, (r,0,2))
plot(dV, (r,0,2))
```
