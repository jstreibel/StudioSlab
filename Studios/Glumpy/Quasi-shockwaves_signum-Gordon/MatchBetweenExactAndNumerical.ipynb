### Teórico

#### Gerando a sequência de zeros

```python
_a0 = 1
_a = [_a0,]

def ξ(k: int) -> float:
    sum = 0.0
    
    for n in range(k):
        an = _a[n]
        sum += ((-1)**(n+k)) * (an**1.5)
    
    akm1=_a[k-1]
    return 8*sum/(akm1**1.5)


def _ak(k: int) -> float:
    if k==0: return _a0
    
    ξ_k = ξ(k)
    
    return .5*_a[k-1]*(1-ξ_k-np.sqrt(1-2*ξ_k))



def generateSequence(N: int, a0: float):
    global _a0, _a
    _a0 = a0
    
    _a = [_a0,]
    for k in range(1,N):
        ak = _ak(k)
        _a.append(ak)
        
    return np.copy(_a)

    
def ck(t, ak:float):
    return np.sqrt(t*t - 4*ak)
```

### Procurando zeros numéricos

#### A forma simples
A forma mais simples de procurar os zeros seria fazer como abaixo: pegar os pontos "perto o suficiente de zero" (usando um treshold) e usando eles. Mas tem um problema que, pra pegar todos os zeros vai ocorrer que, em alguns casos, vão ser pegos dois zeros.

```python
def getIndicesOfZeros(x: np.array, treshold):    
    zeros = np.asarray(np.where(np.abs(x)<treshold))
    phiZeros = phiAt_xeq0[zeros]
    zeros = (zeros[0]/outresT) * _t
    
    return zeros, phiZeros

def plot(ylims, treshold):
    zeros, phiZeros = getIndicesOfZeros(phiAt_xeq0, treshold)
    pyplot.plot((0,_t), (0,0), lw=0.5, color='k')
    pyplot.plot((0,_t), (-treshold,-treshold), lw=0.5, color='orange', label='treshold')
    pyplot.plot((0,_t), (treshold,treshold), lw=0.5, color='orange')
    
    pyplot.scatter(zeros, phiZeros, s=80, marker='o', color='r', label='zeros')
    
    pyplot.plot(tvals, phiAt_xeq0, label='$\phi_\#(t,0)$')
    pyplot.scatter(tvals, phiAt_xeq0, label='$\phi_\#(t,0)$', color='g')
    
    pyplot.ylim(-ylims,ylims)
    pyplot.legend()
    pyplot.axis('off')
    
interact(plot, ylims=(0.1, 0.25, 0.01), treshold=(1e-2, 1e-1, 1e-3))
```

#### A forma que funciona
Pegar cada par de ponto e traçar a reta parametrica que conecta os dois. Se a reta cruzar o eixo 'y', temos um zero. Só.

A reta paramétrica que conecta o ponto $P_1$ com o ponto $P_2$ é descrita por \
$\vec r(t)=P_1 + t(P_2-P_1)=P_1(1-t)+P_2\,t$

O segmento de reta que vai de $\vec r(0)$ até $\vec r(1)$ cruza o eixo $x$ se existir algum $t_0\in\{0,1\}$ para o qual $r_y(t_0)=0$, com $r_y$ sendo a componente $y$ do vetor $\vec r$.

O valor de $t_0$ que zera $r_y$ é tal que $r_y(t_0)=0$. Então, \
$0=P_{1,y} + (P_{2,y}-P_{1,y})\,t_0$, \
\
$t_0=\dfrac{-P_{1,y}}{P_{2,y}-P_{1,y}}$.
\
Para testar se o valor $t_0$ corresponde à algum valor dentro do segmento de reta que conecta os dois pontos, basta testar se $0\leq t_0\leq1$.
\
Voilà
\
Pra simplificar, podemos testar antes se $sign(P_{1,y})==sign(P_{2,y})$
\
\
De uma forma ou de outra, uma vez em posse do $t_0$, agora podemos encontrar o valor no eixo 'x' que corresponde à esse zero:
$x_0 \equiv r_x(t_0) = P_{1,x} + (P_{2,x}-P_{1,x})\,t_0$

```python
f = phiAt_xeq0
n = f.shape[0]

z_t = []
for i in range(n-1):
    P1y, P2y = f[i], f[i+1]
    
    t_0 = -P1y/(P2y-P1y)
    
    if t_0>=0 and t_0<=1:
        P1x, P2x = tvals[i], tvals[i+1]
        z_t.append(P1x+(P2x-P1x)*t_0)
        
z_t = np.array(z_t)
z_zeros = np.zeros(z_t.shape[0])
```

```python
def plot(ylims):        
    pyplot.plot((0,_t), (0,0), lw=0.5, color='k')
    
    pyplot.plot(tvals, phiAt_xeq0, color='g')
    pyplot.scatter(tvals, phiAt_xeq0, label='$\phi_\#(t,0)$', color='g')
    

    pyplot.scatter(z_t, z_zeros, marker='o', color='r', label='zeros')
    
    pyplot.ylim(-ylims,ylims)
    pyplot.legend()
    
    pyplot.axis('off')
    
interact(plot, ylims=(0.01,0.5,0.01))
```
