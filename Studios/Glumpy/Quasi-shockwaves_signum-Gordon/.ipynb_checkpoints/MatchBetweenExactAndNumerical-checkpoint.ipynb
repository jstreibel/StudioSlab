## Visualização de uma seção ao longo de y=0 para três instantes de tempo

```python
def getData(filename: str) -> (np.array, np.array, np.array):
    with open(inputLocation + filename) as myFile:
        firstLine = myFile.readline()

        data = myFile.read().splitlines()
        data = [l.split() for l in data]

        x, fn, fa = [], [], []

        for l in data:
            x.append(float(l[0]))
            fn.append(float(l[1]))
            fa.append(float(l[2]))

        x = np.asarray(x)
        fn = np.asarray(fn)
        fa = np.asarray(fa)
    
    return x, fn, fa


def plotData(x, fn, fa, t0_str, t_str):
    fig = pyplot.figure(figsize=(w,w), dpi=200)
    ax = pyplot.subplot(111)

    label = "$\phi_\#\|_{t=%s}$" % t_str
    ax.plot(x, fn, label=label, zorder=2, lw=0.2, color=numericColor)
    ax.fill(x, fn, zorder=2, color=numericColor, alpha=0.5)
    ax.plot(x, fa, label="$\phi_a\|_{t=%s+%s}$"%(t0_str,t_str),  zorder=3, lw=0.5, color=analyticColor)
    
    ax.set_xlabel("x")
    
    #ax.set_aspect(1)
    ax.set_xlim(-lims,lims)
    #ax.set_ylim(-lims,lims)

    ax.legend()
    
    return ax, fig
    
```

```python
x, fn, fa = getData("file-t0=1.50-t=0.00.dat")
ax,fig1=plotData(x, fn, fa, 't_0', '0')
```

```python
x, fn, fa = getData("file-t0=1.50-t=0.40.dat")
ax,fig2=plotData(x, fn, fa, 't_0', '\Delta t')
```

```python
x, fn, fa = getData("file-t0=1.50-t=0.80.dat")
ax,fig3=plotData(x, fn, fa, 't_0', '2\Delta t')
```

## Zeros da onda de choque

### Teórico

#### Gerando a sequência de zeros

```python
_a0 = 1
_a = [_a0,]

def ξ(k: int) -> float:
    sum = 0.0
    
    for n in range(k):
        an = _a[n]
        sum += ((-1)**(n+k)) * (an**1.5)
    
    akm1=_a[k-1]
    return 8*sum/(akm1**1.5)


def _ak(k: int) -> float:
    if k==0: return _a0
    
    ξ_k = ξ(k)
    
    return .5*_a[k-1]*(1-ξ_k-np.sqrt(1-2*ξ_k))



def generateSequence(N: int, a0: float):
    global _a0, _a
    _a0 = a0
    
    _a = [_a0,]
    for k in range(1,N):
        ak = _ak(k)
        _a.append(ak)
        
    return np.copy(_a)

    
def ck(t, ak:float):
    return np.sqrt(t*t - 4*ak)
```

```python
pyplot.style.use('Solarize_Light2')
```

#### Os $a_k$ são lineares em $a_0$.

Aparentemente. Isso pode ser testado abaixo.

Isso significa que podemos calcular apenas uma vez a sequência dos $a_k$ para um $a_0$ arbitrário, e depois reescalar todos os resultados por $s=a'_0/a_0$. Por conveniência, escolhemos $a_0=1$.

```python
N_a_to_generate = 15
pre_a = generateSequence(N_a_to_generate, 1)
```

```python
def plotAnalytcZeros(a0:float):    
    fig1 = pyplot.figure(figsize=(5,5), dpi=120)
    ax = fig1.add_subplot(111)
    ax.set_xlim(0,10)
    ax.set_ylim(0,10)
    
    a = pre_a*a0 # aparentemente, os a_k são lineares em a0.
    
    for ak in a:
        two = 2.0 + 1.e-7  # pra nao dar bug
        tmin = two*np.sqrt(ak)
        t = np.arange(tmin, 10, .05)

        c = ck(t, ak)
        ax.plot(c,t)
        
    return ax
    
```

###  Sobreposição do resultado numérico com analítico

#### $\phi_\#|_{t=0}=\delta(x)$

```python
fullData00 = openFullData(inputLocation + "swzeros-N2048-t0=0.0.dat")
points00 = fullData00['points']
xData00 = points00[:,0]
tData00 = points00[:,1]
```

```python

def plotAnalytcZerosOnTopOfNumeric(a0:float, t0:float):
    fig1 = pyplot.figure(figsize=(textwidth/2,textwidth/2), dpi=dpi)
    ax = fig1.add_subplot(111)
    ax.set_xlim(0,10)
    ax.set_ylim(0,10)
    
    ax.set_xlabel('r')
    ax.set_ylabel('t')
    
    ax.plot(xData00, tData00, 'b,', label="numeric data")
    
    a = pre_a*a0 # aparentemente, os a_k são lineares em a0.
    
    labelInAlready = False
    for ak in a:
        two = 2.0 + 1.e-7  # pra nao dar bug
        tmin = two*np.sqrt(ak)
        t = np.arange(tmin, 10-t0, .025)

        c = ck(t, ak)
        label=None
        if not labelInAlready:
            label = "analytic data"
            labelInAlready=True
            
        ax.plot(c,t+t0, color='red', ls='-.', lw=0.8, label=label)
        
        
    cone_x = -10, 0, 10
    cone_t = 10, 0, 10
    ax.plot(cone_x, cone_t, color='green', lw=1)
    
    ax.legend()
    
    pyplot.tight_layout()
        
    return fig1, ax
```

#### $\phi_\#|_{t=0}=\phi_a(t_0,x)$

```python
fullData = openFullData(inputLocation + "swzeros-N2048-t0=9.5.dat")
points = fullData['points']
xData = points[:,0]
tData = points[:,1]
```

```python
def plot():
    t0=9.5

    fig1 = pyplot.figure(figsize=(textwidth,textwidth), dpi=dpi)
    ax = fig1.add_subplot(111)
    ax.set_xlim(0,10)
    ax.set_ylim(-t0,10)
    ax.set_aspect(1)
    ax.set_xlabel('r')
    ax.set_ylabel('t')
    
    ax.scatter(xData, tData, label='numeric data', s=0.75)
    #ax.plot(xData, tData, 'b,', alpha=0.25, label='numeric data')

    a = generateSequence(40, 1)

    labelDone = False
    for ak in a:
        two = 2.0 + 1.e-7  # pra nao dar bug
        tmin = two*np.sqrt(ak)
        t = np.arange(tmin, 10+t0, .025)

        c = ck(t, ak)
        
        label = None
        if not labelDone:
            label="analytic zeros"
            labelDone=True
            
        ax.plot( c,t-t0, '-.', color='k', lw=0.5, label=label)
        ax.plot(-c,t-t0, '-.', color='k', lw=0.5)


    cone_x = (-10-t0, 0-t0, 10-t0), (-10+t0, 0+t0, 10+t0)
    cone_t = 10, 0, 10
    ax.plot(cone_x[0], cone_t, color='green', lw=1, label='lightcone of border')
    ax.plot(cone_x[1], cone_t, color='green', lw=1)
    
    ax.plot((0,10), (-t0,0), color='red', lw=1, label='lightcone of analytic shock-wave')
    
    ax.plot((-10,10), (0,0), lw=1, color='b', alpha=0.5, label="begin simulation")
    
    ax.legend()
    
    pyplot.tight_layout()
    
    return fig1, ax
    
```

```python
fig, ax = plot()
```
