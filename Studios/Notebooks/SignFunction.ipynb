```julia
function logz(z_re, z_im)
    magnitude = sqrt(z_re^2 + z_im^2)
    angle = atan(z_im/z_re) # o certo aqui eh atan2
    return log(magnitude) + 1im * angle
end

function tanh⁻¹(z)
    ℜz = real(z)
    ℑz = imag(z)
    return ((logz(1+ℜz, ℑz))-logz(1-ℜz, ℑz))/2
end
```
