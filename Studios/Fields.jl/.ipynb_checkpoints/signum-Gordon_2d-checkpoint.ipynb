```julia
using DifferentialEquations, CUDA
```

```julia
using Fields
```

# Parameters

```julia
L = 2
N = 1024
tmax = 1.0

n = 256
tsave = range(0.0, stop=tmax, length=n)
frac = Float32(N)/n
rsave_idxs = [Integer(i*frac + (j-1)*frac*N) for i in 1:n, j in 1:n]
;
```

### Total saved values:

```julia
MB =  (n^3 * 8) * 10^-6
print("$MB MB")
```

```julia
h = L/(N-1)
x₀ = (L/2, L/2)
x = [((i-1)h, (j-1)h).-x₀ for j in 1:N, i in 1:N]
r = [sqrt(x^2+y^2) for (x,y) in x]
t = 0.0:h/10:tmax
;
```

# Initial conditions + allocs.

### Clean mem

```julia
sol = nothing
GC.gc()

CUDA.reclaim()
```

### Setup IC

```julia
φ₀   = sin(r)
∂ₜφ₀ = zero(r)
uₕₒₛₜ = cat(φ₀, ∂ₜφ₀; dims=3)
u₀ = CuArray(uₕₒₛₜ)
∇²ϕ  = CuArray(zero(r))
;
```

### View IC

```julia
using Plots; gr()

φ₀_plot   = plot(φ₀,  st=:surface)
∂ₜφ₀_plot = plot(∂ₜφ₀,st=:surface)

plot(φ₀_plot, ∂ₜφ₀_plot, layout=(1, 2), size=(1300, 400))
```

# Problem

```julia
function signumGordon_2d!(uₜ, u, (N, h, ∇²ϕ), t)
    ϕ  = @views u[:,:,1]
    ϕₜ = @views u[:,:,2]

    compute_∇²_2d!(∇²ϕ, ϕ, h)

    # Eq. system
    # φ ≡ ϕₜ
    # φ ≡ ϕₜₜ
    φ  = ϕₜ
    φₜ = ∇²ϕ - sign.(ϕ)

    # Output
    uₜ[:,:,1] = φ
    uₜ[:,:,2] = φₜ

    nothing
end

prob = ODEProblem(signumGordon_2d!, u₀, (0.0, tmax), (N, h, ∇²ϕ))

;
```

# Solution

```julia
@time sol = solve(prob, RK4(); saveat=tsave, save_idxs=rsave_idxs, adaptive=false, dt=h/10)
;
```

### Organize data

```julia
matrix_array = Array{Matrix{Float64}}(undef, n)

ϕₕₒₛₜ   = [Matrix{Float64}(undef, n, n) for _ in 1:n]
∂ₜϕₕₒₛₜ = [Matrix{Float64}(undef, n, n) for _ in 1:n]


# Populate the Array with n x n matrices
for i in 1:n
    matrix_array[i] = Matrix{Float64}(undef, n, n)
end

for i ∈ 1:n
    copyto!(uₕₒₛₜ, sol.u[1])
end
ϕ  = uₕₒₛₜ[:,:,1]
ϕₜ = uₕₒₛₜ[:,:,2]

;
```
