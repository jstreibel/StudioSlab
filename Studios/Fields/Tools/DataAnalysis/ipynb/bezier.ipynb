```python
start_value = 0
end_value = 1
a = start_value
b = end_value
Δ = b-a

k1=1
k2=1

P0 = a
P1 = a + Δ*k1
P2 = a + Δ*k2
P3 = b

def bezier_curve(t):
    global P0, P1, P2, P3
    
    return (1 - t)**3*P0  +  3*t*(1 - t)**2*P1  +  3 * (1 - t) * t**2*P2  +  t**3*P3


    
```

```python
plotting.plot(bezier_curve(t),         xlim=(0,1), ylim=(0,3),   adaptive=False, nb_of_points=1000)
plotting.plot(bezier_curve(t).diff(t), xlim=(0,1), ylim=(0,3.1), adaptive=False, nb_of_points=1000)
```

```python
from matplotlib import pyplot as plt
import numpy as np
import scipy
```

```python
Γ = scipy.special.gamma
```

```python
def C(m,n):
    a = (n+m)/2
    b = (n-m)/2
    
    return (Γ(n+1)/(Γ(1+a)*Γ(1+b)))/(2**n)
```

```python
m_min = 1
m_range = 7
for m in range(m_min, m_range+1, 2):
    n_min = m
    n_max = 30
    
    x1 = np.arange(n_min, n_max, 2)
    x2 = np.linspace(n_min, n_max, 100)
    
    plt.scatter(x1, C(m, x1), marker='o', label='m='+str(m))
    plt.plot(x2, C(m, x2))

plt.xlabel('n')    
plt.legend()
```

```python
n_min = 1
n_range = 21
for n in range(n_min, n_range+1, 2):
    m_min = -n
    m_max = n
    
    x1 = np.arange(m_min, m_max+1, 2)
    x2 = np.linspace(m_min, m_max, 100)
    
    plt.scatter(x1, C(x1, n), marker='o', label='n='+str(n))
    plt.plot(x2, C(x2, n))

plt.xlabel("m")    
plt.legend()
```

```python
n = np.arange(1, 15, 0.1)
m = np.arange(1, 15, 0.1)
X, Y = np.meshgrid(n, m)

# Define the density function
Z = C(X, Y)

# Create the plot
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(label='Density')
plt.xlabel('n')
plt.ylabel('m')
plt.title('2D Density Function')
plt.show()
```

```python
def P(m, n):
    return C(m, n)/C(m, n-2)
```

```python

for m in np.arange(1, 11, 1):
    n = np.arange(m+2, 25, 0.5)
    
    plt.plot(n, P(m, n), color='grey', linewidth=0.5, alpha=1)
```
