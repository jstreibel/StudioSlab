```python
from sympy import *
init_printing()
```

```python
t = Symbol('t')
```

```python
start_value = 0
end_value = 1
a = start_value
b = end_value
Δ = b-a

k1=1
k2=1

P0 = a
P1 = a + Δ*k1
P2 = a + Δ*k2
P3 = b

def bezier_curve(t):
    global P0, P1, P2, P3
    
    return (1 - t)**3*P0  +  3*t*(1 - t)**2*P1  +  3 * (1 - t) * t**2*P2  +  t**3*P3


    
```

```python
plotting.plot(bezier_curve(t),         xlim=(0,1), ylim=(0,3),   adaptive=False, nb_of_points=1000)
plotting.plot(bezier_curve(t).diff(t), xlim=(0,1), ylim=(0,3.1), adaptive=False, nb_of_points=1000)
```
