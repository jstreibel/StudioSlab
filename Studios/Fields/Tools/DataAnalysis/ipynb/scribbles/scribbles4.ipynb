```python
from sympy import *
init_printing()
```

```python
x = Symbol('x', real=True)
```

```python
def O(phi, n):
    phi_sym = Integer(0)
    
    for i in range(1, 2*n, 2):
        phi_sym += 1/i * phi.subs(x, i*x)
        
    return 4/pi * phi_sym
        
```

```python
phi = Function('phi')
phi
```

```python
from numpy import linspace
for a in linspace(0, 1.0, 10):
    plot(O(sin(x+a*pi), 25))
```
