```python
from sympy import *

N = Symbol('N', positive=True, odd=True)
n = Symbol('n', positive=True, odd=True)
z = Symbol('z')
A = Symbol('A')
```

```python
Stirling = lambda z: (z-Rational(1,2))*log(z) - z + Rational(1,2)*log(2*pi)
display(Stirling(z))
```

```python
npN = (N+n)/2+1
nmN = (N-n)/2+1
np1 = n+1
```

```python
func_a = lambda z: log(gamma(z))
func_b = lambda z: loggamma(z)
func_c = lambda z: Stirling(z)

term_fatidique = n*log(*(2/A)) + log(2/n/pi)

def coeff(func):
    return (func(npN) - func(nmN) - func(np1) + term_fatidique)
```

```python
coeff_a = 2/N*coeff(func_a)
display(coeff_a)

coeff_c = 2/N*coeff(func_c)
display(coeff_c)
coeff_c1 = 2/N - (1-n/N+1/N)*log((N-n)/2+1) + (1+n/N+1/N)*log((N+n)/2+1) - (2*n+1)/N*log(n+1) - 1/N*log(2*pi) + 2/N*term_fatidique
display(coeff_c1)
coeff_c2 = 2/N - (1-n/N+1/N)*log((N-n)/2+1) + (1+n/N+1/N)*log((N+n)/2+1) - (2*n+1)/N*log(n+1) - 1/N*log(2*pi) + 2/N*term_fatidique
display(coeff_c2)

coeff_d = - (1-n/N)*log((N-n)/2) + (1+n/N)*log((N+n)/2) - (2*n)/N*log(n+1) + 2/N*term_fatidique
display(coeff_d)
coeff_d2 = (n/N-1)*log((N/2)*(1-n/N)) + (n/N+1)*log((N/2)*(1+n/N)) - (2*n)/N*log(n+1) + 2/N*term_fatidique
display(coeff_d2)
coeff_d3 = n/N*(log((N/2)*(1-n/N))+log((N/2)*(1+n/N))) -(log((N/2)*(1-n/N))-log((N/2)*(1+n/N))) - (2*n)/N*log(n+1) + 2/N*term_fatidique
display(coeff_d3)
coeff_d4 = n/N*log((N**2/4)*(1+n/N)*(1-n/N)) + log((1+n/N)/(1-n/N)) - (2*n)/N*log(n+1) + 2/N*term_fatidique
display(coeff_d4)
coeff_d5 = n/N*log((1+n/N)*(1-n/N)) + log((1+n/N)/(1-n/N)) - (2*n)/N*log((n+1)*2/N) + 2/N*term_fatidique
display(coeff_d5)

coeff_e = n/N*log((1+n/N)*(1-n/N)) + log((1+n/N)/(1-n/N)) - (2*n)/N*log(2*n/N)
display(coeff_e)
```

```python
N_num = 55
A_num = 2

#func = exp(N/2*z)/(n+1)**5
func = z

coeff_a_num   = func.subs(z, coeff_a).subs(N, N_num).subs(A, A_num)
#coeff_c_num  = func.subs(z, coeff_c).subs(N, N_num).subs(A, A_num)
#coeff_c1_num = func.subs(z, coeff_c1).subs(N, N_num).subs(A, A_num)
coeff_c2_num  = func.subs(z, coeff_c2).subs(N, N_num).subs(A, A_num)
#coeff_d_num  = func.subs(z, coeff_d).subs(N, N_num).subs(A, A_num)
#coeff_d2_num = func.subs(z, coeff_d2).subs(N, N_num).subs(A, A_num)
#coeff_d3_num = func.subs(z, coeff_d3).subs(N, N_num).subs(A, A_num)
coeff_d4_num  = func.subs(z, coeff_d4).subs(N, N_num).subs(A, A_num)
coeff_e_num   = func.subs(z, coeff_e).subs(N, N_num).subs(A, A_num)

plot(coeff_a_num, coeff_c2_num, coeff_d4_num, coeff_e_num, (n, 2, N_num-1))
```
