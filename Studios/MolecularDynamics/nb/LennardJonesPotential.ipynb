```python
from sympy import *
init_printing()
```

```python
r, alpha, eps, sigma = symbols("r alpha epsilon sigma")
```

```python
#auto inv_n = σ / r;
#return 4. * ε * (POW12(inv_n) - POW6(inv_n));

U = lambda r: 4/eps * ((sigma/r)**12 - (sigma/r)**6 ) 
U(r)
```

```python
p = ((1,1),(0.27,36))[0]

V = U(r).subs(((sigma, p[0]), (eps, p[1])))
temp2 = -V.diff(r)
plot(V, (r,p[0]-0.01,2*p[0]))
plot(temp2, (r,p[0]-0.01,2*p[0]))

temp2
```
